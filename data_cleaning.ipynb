##### Fix height and weight plus add bmi

```python
def clean_height(height):
    # If the height is null (NaN), return NaN
    if pd.isna(height):
        return float('nan')
    # If the height is in centimeters (cm), extract the numeric value before 'cm' and convert it to float
    elif 'cm' in height:
        height_cm = float(height.split('cm')[0].strip())
        # Round up the height to the nearest integer using math.ceil function
        return math.ceil(height_cm)
    # If the height is in inches (in), extract the numeric value before 'in' and convert it to float
    elif 'in' in height:
        height_in = float(height.split('in')[0].strip())
        # Convert the height from inches to centimeters
        height_cm = height_in * 2.54
        # Round up the height to the nearest integer using math.ceil function
        return math.ceil(height_cm)
    else:
        return float('nan')

# Apply the clean_height function to the 'height' column to clean and convert the values
df_games_conso['height'] = df_games_conso['height'].apply(clean_height)

# Replace values smaller than 150 with 150 and values higher than 199 with 199 in the 'height' column
df_games_conso['height'] = np.clip(df_games_conso['height'], 150, 199)

def clean_weight(weight):
    # If the weight is null (NaN), return NaN
    if pd.isna(weight):
        return float('nan')
    # If the weight is in kilograms (kg), extract the numeric value before 'kg' and convert it to float
    elif 'kg' in weight:
        weight_kg = float(weight.split('kg')[0].strip())
        # Round up the weight to the nearest integer using math.ceil function
        return math.ceil(weight_kg)
    # If the weight is in pounds (lb), extract the numeric value before 'lb' and convert it to float
    elif 'lb' in weight:
        weight_lb = float(weight.split('lb')[0].strip())
        # Convert the weight from pounds to kilograms
        weight_kg = weight_lb / 2.205
        # Round up the weight to the nearest integer using math.ceil function
        return math.ceil(weight_kg)
    else:
        return float('nan')

# Apply the clean_weight function to the 'weight' column to clean and convert the values
df_games_conso['weight'] = df_games_conso['weight'].apply(clean_weight)

# Replace values smaller than 50 with 50 and values higher than 109 with 109 in the 'weight' column
df_games_conso['weight'] = np.clip(df_games_conso['weight'], 50, 109)

# Calculate BMI (Body Mass Index) while converting height from cm to m (BMI = weight (kg) / (height (m) ^ 2)
df_games_conso['bmi'] = df_games_conso['weight'] / ((df_games_conso['height'] / 100) ** 2)
```

##### Fix overallRank and overallScore

```python
# Replace 'DQ' (Disqualified) with -1, indicating the competitor was disqualified
df_games_conso['overallRank'] = df_games_conso['overallRank'].replace("DQ", -1)
# Convert 'overallRank' to integer data type
df_games_conso['overallRank'] = df_games_conso['overallRank'].astype(np.int64)

# Define a dictionary 'conditions' to map specific competitorId and year to their corresponding overallRank values
conditions = {
    (455677, 2021): 38,
    (592472, 2021): 39,
    (1019212, 2021): 40,
    (901702, 2021): 39,
    (975774, 2021): 40,
    (975774, 2018): 40,
    (3200, 2015): 40,
    (14354, 2014): 40,
    (151906, 2013): 47
}

# Use the 'conditions' dictionary to map the 'overallRank' column based on the 'competitorId' and 'year' conditions
df_games_conso['overallRank'] = df_games_conso.apply(lambda row: conditions.get((row['competitorId'], row['year']), row['overallRank']), axis=1)

# Replace ':' with an empty string in 'overallScore' column and divide by 100 only for rows where ':' is present
mask = df_games_conso['overallScore'].str.contains(':')
df_games_conso.loc[mask, 'overallScore'] = df_games_conso.loc[mask, 'overallScore'].str.replace(':', '').astype(float) / 100

# Convert 'overallScore' to integer data type
df_games_conso['overallScore'] = df_games_conso['overallScore'].astype(np.int64)
```

##### Replace 0 from age and null from height, weight, bmi and affiliateId

```python
# Create the 'ageNull' column and fill it with 1 if 'age' is equal to 0, 0 otherwise
df_games_conso['ageNull'] = (df_games_conso['age'] == 0).astype(np.int64)

# Calculate the average age of non-zero values and replace 0 values with the average
average_age = df_games_conso.loc[df_games_conso['age'] != 0, 'age'].mean()
df_games_conso['age'] = df_games_conso['age'].replace(0, average_age).astype(np.int64)

# Create the 'heightNull', 'weightNull', and 'bmiNull' columns and fill them with 1 if the corresponding value is null, 0 otherwise
df_games_conso['heightNull'] = df_games_conso['height'].isnull().astype(np.int64)
df_games_conso['weightNull'] = df_games_conso['weight'].isnull().astype(np.int64)
df_games_conso['bmiNull'] = df_games_conso['bmi'].isnull().astype(np.int64)

# Calculate the average height, weight, and bmi broken down by gender
average_height_by_gender = df_games_conso.groupby('gender')['height'].mean()
average_weight_by_gender = df_games_conso.groupby('gender')['weight'].mean()
average_bmi_by_gender = df_games_conso.groupby('gender')['bmi'].mean()

# Fill null values in 'height', 'weight', and 'bmi' with the respective average values based on gender
df_games_conso['height'] = df_games_conso.apply(lambda row: average_height_by_gender[row['gender']] if pd.isnull(row['height']) else row['height'], axis=1)
df_games_conso['weight'] = df_games_conso.apply(lambda row: average_weight_by_gender[row['gender']] if pd.isnull(row['weight']) else row['weight'], axis=1)
df_games_conso['bmi'] = df_games_conso.apply(lambda row: average_bmi_by_gender[row['gender']] if pd.isnull(row['bmi']) else row['bmi'], axis=1)

# Fill null values in 'affiliateId' with 0 and convert 'affiliateId' to integer data type
df_games_conso['affiliateId'] = df_games_conso['affiliateId'].fillna(0).astype(np.int64)
```

##### Add gamesCompetitions

```python
# Sort the DataFrame by 'competitorId' and 'year' in ascending order
df_games_conso.sort_values(by=['competitorId', 'year'], inplace=True)

def count_competitions(df, mask_column, new_column):
    # Create a mask to filter rows based on the provided column
    mask = df[mask_column] > 0

    # Convert the mask to 1s and 0s using astype(int)
    mask = mask.astype(int)

    # Group the DataFrame by 'competitorId' and apply the cumsum function to count the occurrences for each competitorId
    df[new_column] = mask.groupby(df['competitorId']).cumsum()

    # Replace the 0 values with NaN to differentiate from other years without competitions
    df[new_column].replace(0, np.nan, inplace=True)

    # Forward fill the null values for each competitorId separately
    df[new_column] = df.groupby('competitorId')[new_column].fillna(method='ffill')

    # Replace any remaining null values with 0 and convert to integer data type
    df[new_column].fillna(0, inplace=True)
    df[new_column] = df[new_column].astype(np.int64)

# Call the function for 'gamesCompetitions' column
count_competitions(df_games_conso, 'overallScore', 'gamesCompetitions')
```
