# Data Cleaning

### Games

```python
df_games_conso = pd.read_csv(base_path + "csv_files/version_1_raw_data/df_games_conso.csv")

df_games_conso.loc[df_games_conso['countryOfOriginName'] == 'Namibia', 'countryOfOriginCode'] = 'NA'

df_games_conso.loc[df_games_conso['countryOfOriginCode'] == 'BY', 'countryOfOriginName'] = 'Belarus'

df_games_conso.loc[df_games_conso['countryOfOriginCode'] == 'RU', 'countryOfOriginName'] = 'Russian Federation'

asia_countries = ['RU', 'AE']
df_games_conso.loc[df_games_conso['countryOfOriginCode'].isin(asia_countries), 'regionName'] = 'Asia'

europe_countries = ['FI', 'FR', 'DE', 'HU', 'IS', 'NO', 'SE', 'CH']
df_games_conso.loc[df_games_conso['countryOfOriginCode'].isin(europe_countries), 'regionName'] = 'Europe'

north_america_countries = ['US', 'CA']
df_games_conso.loc[df_games_conso['countryOfOriginCode'].isin(north_america_countries), 'regionName'] = 'North America'

oceania_countries = ['AU', 'NZ']
df_games_conso.loc[df_games_conso['countryOfOriginCode'].isin(oceania_countries), 'regionName'] = 'Oceania'

df_games_conso = df_games_conso.drop(columns=['regionId'])
```

```python
# Group by 'countryOfOriginName' and 'countryOfOriginCode' and count the number of unique 'regionName' values for each country
region_counts = df_games_conso.groupby(['countryOfOriginName', 'countryOfOriginCode'])['regionName'].nunique()

# Filter countries with two or more unique 'regionName' values
countries_with_multiple_regions = region_counts[region_counts >= 2]

# Print the result
countries_with_multiple_regions
```

```python
# Filter the DataFrame to include only rows where 'countryOfOriginName' is 'Australia'
australia_df = df_games_conso[df_games_conso['countryOfOriginName'] == 'United States']

# Get the different values of 'regionName' for Australia
unique_regions_australia = australia_df['regionName'].unique()

# Print the result
unique_regions_australia
```

```python
# Filter the DataFrame to include only rows where 'countryOfOriginName' is 'Australia'
australia_df = df_games_conso[(df_games_conso['regionName'].notnull()) & (df_games_conso['countryOfOriginName'].isnull())]

# Get the different values of 'regionName' for Australia
unique_regions_australia = australia_df['regionName'].unique()

# Print the result
unique_regions_australia
```
