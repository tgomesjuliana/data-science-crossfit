# Data Cleaning

### Games

```python
df_games_conso = pd.read_csv(base_path + "csv_files/version_1_raw_data/df_games_conso.csv")

df_games_conso.loc[df_games_conso['countryOfOriginName'] == 'Namibia', 'countryOfOriginCode'] = 'NA'

df_games_conso.loc[df_games_conso['countryOfOriginCode'] == 'BY', 'countryOfOriginName'] = 'Belarus'

df_games_conso.loc[df_games_conso['countryOfOriginCode'] == 'RU', 'countryOfOriginName'] = 'Federation'
```

```python
df_games_conso
```

```python
df_games_conso.isnull().sum()
```

```python
# Group by 'countryOfOriginName' and count the number of unique 'regionName' values for each country
region_counts = df_games_conso.groupby('countryOfOriginName')['regionName'].nunique()

# Filter countries with two or more unique 'regionName' values
countries_with_multiple_regions = region_counts[region_counts >= 2]

# Print the result
countries_with_multiple_regions
```

```python
# Filter the DataFrame to include only rows where 'countryOfOriginName' is 'Australia'
australia_df = df_games_conso[df_games_conso['countryOfOriginName'] == 'Iceland']

# Get the different values of 'regionName' for Australia
unique_regions_australia = australia_df['regionName'].unique()

# Print the result
unique_regions_australia
```

```python
df = df_games_conso[(df_games_conso['countryOfOriginName'].notnull()) & (df_games_conso['regionName'].isnull())]

df

# df['countryOfOriginName'].unique()
```
